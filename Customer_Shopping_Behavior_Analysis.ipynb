# Finding age groups to marketting and understanding customer behavior

```python
# Creates A new Column age_group
labels=['Young Adult','Adult','Middle-aged','Senior']#Each element is a label (category name)
df['age_group'] = pd.qcut(df['age'], q=4, labels= labels)  
```

##### pd.qcut() divides the data into equal-sized groups

qcut = Quantile Cut

q=4 means split the ages into 4 equal-sized groups (25% each)

Example:

Group 1 → Lowest 25% ages

Group 2 → Next 25%

Group 3 → Next 25%

Group 4 → Highest 25%

So it divides your ages into 4 categories based on distribution — not based on fixed age ranges.

Assign readable labels instead of numbers

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']


Instead of numeric bins like 0,1,2,3 → you get meaningful names.

# Create column purchase_frequency_days To find when customer buys items


```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 1,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' :90,
    'Bi-Weekly' :14,
    'Annually' : 365,
    'Every 3 monnths' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

### What .map() Does in Your Code
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

✔️ .map() looks at each value in the column

✔️ Finds its matching value in your dictionary (frequency_mapping)

✔️ Replaces it with the mapped number

✔️ Stores the result in a new column

#### Summary
✔ It creates a new column (purchase_frequency_days)

✔ It does NOT replace the original column (frequency_of_purchases)

✔ .map() converts text → numbers using your dictionary
